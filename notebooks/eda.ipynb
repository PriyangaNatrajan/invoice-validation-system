```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

```python
df = pd.read_csv("../data/invoices.csv")
df.head()
```

```python
print("shape of dataset:",df.shape())
print("\ndata types:\n",df.dtypes)
```

```python
print("missing values:\n")
print(df.isnull().sum())
```

```python
print("Duplicate rows:", df.duplicated().sum())
```

```python
df['invoice_date'] = pd.to_datetime(df['invoice_date'], errors='coerce')
df['due_date'] = pd.to_datetime(df['due_date'], errors='coerce')
```

```python
df.describe()
```

```python
num_colmns=['qty','unit_price','subtotal','tax_amount','total_amount']
print("negative values:\n")
print((df[num_colmns]<0).sum())
```

```python
invalid_dates=df[df['due_date']<df['invoice_date']]
print("rows with due_date earlier than invoice_date:",len(invalid_dates))
invalid_dates
```

```python
for col in num_colmns:
    plt.figure()
    sns.histplot(df[col],kde=True)
    plt.title(f"{col} Distribution")
    plt.show
```

```python
for col in num_colmns:
    plt.figure()
    sns.boxplot(x=df[col])
    plt.title(f"{col} Boxplot")
    plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df[num_colmns].corr())
plt.title("Correlation Matrix")
plt.show()
```

```python
invalid_total=df[df['total_amount'] < df['subtotal']]
print("invoices where total<subtotal:",len(invalid_total))
invalid_total.head()
```
